```python
import numpy as np
import pandas as pd

def make_df(cenfile):
    df=pd.read_csv(cenfile, names=['num','cenx','ceny','fwhm','peak','flux'], delim_whitespace=True)
    return df
def make_df2(cenfile):
    df=pd.read_csv(cenfile, names=['num','cenx','ceny','fwhm','peak','flux','maxadu'], delim_whitespace=True)
    return df
def make_df_from_pos(posfile):
    # 00003166.714 00002155.418 013.492 0020 002.243
    df=pd.read_csv(posfile, names=['cenx','ceny','mag','num','fwhm'], delim_whitespace=True)
    return df
def find_nearest_element(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx
def find_nearest_point(xvals,yvals, xvalue,yvalue):
    x= np.asarray(xvals)
    y= np.asarray(yvals)
    idx=np.sqrt((x-xvalue)**2+(y-yvalue)**2)
    return idx.argmin(),idx[idx.argmin()]

def locate(df1,df2,cen_num):
    '''
    For a given centroid numer in dataframe df1 find the closest match in
    dataframe df2 
    '''
    xvalue=float(df1.loc[df1['num'] == cen_num]['cenx'])
    yvalue=float(df1.loc[df1['num'] == cen_num]['ceny'])
    #print(xvalue,yvalue)
    xvals=df2['cenx']
    yvals=df2['ceny']
    idx=find_nearest_point(xvals,yvals, xvalue,yvalue)
    #print(idx)
    cen_num_df2=df2.iloc[[idx]]['num'] 
    return cen_num_df2

def distance(df1,num1,df2,num2):
    '''
    calculates the distance between two centroits in the two dataframes df1 and df2
    '''
    xvalue1=float(df1.loc[df1['num'] == num1]['cenx'])
    yvalue1=float(df1.loc[df1['num'] == num1]['ceny'])
    
    xvalue2=float(df2.loc[df2['num'] == num2]['cenx'])
    yvalue2=float(df2.loc[df2['num'] == num2]['ceny'])
    return np.sqrt((xvalue1-xvalue2)**2+(yvalue1-yvalue2)**2),xvalue1-xvalue2,yvalue1-yvalue2
```
